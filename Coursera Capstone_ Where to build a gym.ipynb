### Introduction to the problem

A Lansing Michigan investor is looking to open up a new gym in town to compete with Planet Fitness, while not being too close to other existing gyms. As a contracted Data Scientist I will use the Foursquare API to gather data about locations of local gyms; as well as populations in those areas to determine the best location for a new gym to open. The investors target audience is people that live too far for other gyms to be convenient, as well as people that can not afford some of the pricier gyms in their area. 

### Data

The bulk of my data will be coming from the Foursquare API. First I will define an area where the business owner has an idea of opening a gym, and then use the Foursquare API data to see its vicinity to other gyms and health clubs within a given radius. I will also be gathering data from the US Census Bureau to find data about the population in the surrounding cities. Is there a city that is more elderly? Richer? Poorer? What is actually important enought to take into consideration? All of the data is integer based and does not seem to be missing any values so it should be easy to work with. After all the information is gathered I will pick the best general area around the city of Lansing for my client to open their gym.
